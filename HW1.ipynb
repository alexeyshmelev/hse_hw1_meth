# Main work

```python
import subprocess
import multiprocessing
import matplotlib.pyplot as plt
import numpy as np
```

```python
!samtools view ./cell8/SRR5836473_1_bismark_bt2_pe.bam | head
```

```python
input_file = ["./cell8/SRR5836473_1_bismark_bt2_pe.bam", "./ICM/SRR5836475_1_bismark_bt2_pe.bam", "./epiblast/SRR3824222_1_bismark_bt2_pe.bam"]
output_file = ["sorted_8cell.bam", "sorted_icm.bam", "sorted_epiblast.bam"]

def sort_files(input_file, output_file):
    subprocess.call(["samtools", "sort",  input_file, "-o", output_file])
    
p1 = multiprocessing.Process(target=sort_files, args=(input_file[0], output_file[0]))
p2 = multiprocessing.Process(target=sort_files, args=(input_file[1], output_file[1]))
p3 = multiprocessing.Process(target=sort_files, args=(input_file[2], output_file[2]))

p1.start()
p2.start()
p3.start()

p1.join()
p2.join()
p3.join()
```

```python
input_file = ["./cell8/SRR5836473_1_bismark_bt2_pe.bam", "./ICM/SRR5836475_1_bismark_bt2_pe.bam", "./epiblast/SRR3824222_1_bismark_bt2_pe.bam"]
output_file = ["s_8_cell", "s_icm", "s_epiblast"]

def deduplicate(input_file, output_file):
    subprocess.call(["deduplicate_bismark", "--bam",  "--paired", "-o", output_file, input_file])

p1 = multiprocessing.Process(target=deduplicate, args=(input_file[0], output_file[0]))
p2 = multiprocessing.Process(target=deduplicate, args=(input_file[1], output_file[1]))
p3 = multiprocessing.Process(target=deduplicate, args=(input_file[2], output_file[2]))

p1.start()
p2.start()
p3.start()

p1.join()
p2.join()
p3.join()
```

```python
input_file = ["s_8_cell.deduplicated.bam", "s_epiblast.deduplicated.bam", "s_icm.deduplicated.bam"]

def methylation_extractor(input_file):
    subprocess.call(["bismark_methylation_extractor", "--gzip",  "--bedGraph", "--cytosine_report", "--genome_folder", "./mouse_chr", input_file])
    
p1 = multiprocessing.Process(target=methylation_extractor, args=(input_file[0],))
p2 = multiprocessing.Process(target=methylation_extractor, args=(input_file[1],))
p3 = multiprocessing.Process(target=methylation_extractor, args=(input_file[2],))

p1.start()
p2.start()
p3.start()

p1.join()
p2.join()
p3.join()
```

```python
files = ["s_8_cell.deduplicated.bedGraph", "s_epiblast.deduplicated.bedGraph", "s_icm.deduplicated.bedGraph"]
for file in files:
    with open(file, "r") as f:
        data = []
        for line in f:
            if line.split("\t")[0] == "chr11":
                data.append(float(line.split("\t")[3][:-1]))
        names, frequency = np.unique(data, return_counts=True)
        plt.clf()
        plt.xlabel("percentage")
        plt.ylabel("frequency")
        xn = [f"{(i)*10}" for i in range(11)]
        xp = np.arange(0, len(frequency)+1, len(frequency)/10)
        plt.xticks(xp, xn)
        plt.title(file)
        plt.plot(frequency/len(data))
        plt.show()
```
